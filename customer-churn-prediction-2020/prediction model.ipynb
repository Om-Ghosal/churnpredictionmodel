```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression,LogisticRegression
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
le=LabelEncoder()
sc = StandardScaler()
data= pd.read_csv('train.csv')
data
```

```python
list=['state','area_code','international_plan','voice_mail_plan','churn']
for l in list :
    temp = le.fit_transform(data[l])
    data.drop(l,axis=1,inplace=True)
    data[l]=temp
data.head(5)
```

```python
a=data.drop(['churn'],axis=1)
X=sc.fit_transform(a)
y=data['churn']

data
```

```python
data_test = pd.read_csv("test.csv")


list=['state','area_code','international_plan','voice_mail_plan']
for l in list :
    temp = le.fit_transform(data_test[l])
    data_test.drop(l,axis=1,inplace=True)
    data_test[l]=temp
data_test.head(5)

X_test = sc.fit_transform(data_test.drop(['id'],axis=1))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
#from sklearn.preproccesing import LinearScaler

o= LogisticRegression()
o.fit(X,y)
o.predict(X_test)

samp = pd.read_csv('sampleSubmission.csv')
list=['churn']
for l in list :
    temp = le.fit_transform(samp[l])
    samp.drop(l,axis=1,inplace=True)
    samp[l]=temp
samp.head(5)


y_test=samp.drop(['id'],axis=1)
o.score(X_test,y_test)
```
